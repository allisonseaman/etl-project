```python
import pandas as pd
from sqlalchemy import create_engine

import numpy as np
```

```python
new_york_data=pd.read_csv("new_york_crimes.csv")
chicago_data=pd.read_csv("chicago_crimes.csv")
```

```python
new_york_data.head()
```

```python
#choosing columns we need
chicago_data = chicago_data.drop(["Case Number", "Block", "Location Description", "Beat", "District", "Ward", "Community Area", "FBI Code", "X Coordinate", "Y Coordinate","Latitude","Longitude","Updated On","IUCR"], axis = 1)
chicago_data.head()
```

```python
new_york_data=new_york_data.drop(["CMPLNT_TO_DT","CMPLNT_TO_TM","CMPLNT_FR_TM",  "ADDR_PCT_CD", "RPT_DT", "PD_CD","X_COORD_CD","Y_COORD_CD",
                                  "JURISDICTION_CODE", "Latitude","Longitude","KY_CD","HOUSING_PSA","HADEVELOPT","LOC_OF_OCCUR_DESC", "JURIS_DESC", "PARKS_NM","PATROL_BORO"], axis=1)
new_york_data.head()
```

```python
#renaming columns
new_york_data = new_york_data.rename(columns={"CMPLNT_NUM":"ID","CMPLNT_FR_DT":"Date","CMPLNT_FR_TM":"Time",
                                              "ADDR_PCT_CD":"Precinct of Incident","RPT_DT":"Reported Date",
                                              "OFNS_DESC":"Primary Type","PD_DESC":"Description","CRM_ATPT_CPTD_CD":"Crime Status",
                                              "LAW_CAT_CD":"Offense Level","BORO_NM":"Borough","LOC_OF_OCCUR_DESC":"Specific Location",
                                              "PREM_TYP_DESC":"Description of Prem","JURIS_DESC":"Juris Desc",
                                              "JURISDICTION_CODE":"Juris Code","SUSP_AGE_GROUP":"Suspect Age","SUSP_RACE":"Suspect Race",
                                              "SUSP_SEX":"Suspect Sex","TRANSIT_DISTRICT":"Transit District","Lat_Lon":"Geo Location",
                                              "STATION_NAME":"Transit Station","VIC_AGE_GROUP":"Victim Age","VIC_RACE":"Victim Race",
                                              "VIC_SEX":"Victim Sex"})
new_york_data.head()
```

```python
new_york_data=new_york_data.drop(["Crime Status","Offense Level","Borough","Description of Prem","Suspect Age",
                                  "Suspect Race","Suspect Sex","Transit District","Geo Location","Transit Station", 
                                  "Victim Age","Victim Race","Victim Sex"], axis=1)
new_york_data.head()
```

```python
chicago_data=chicago_data.drop(["Arrest", "Domestic", "Year", "Location"], axis=1)
chicago_data.head()
```

```python
start_date='1/1/2006'
end_date='12/31/2017'
new2 = (new_york_data['Date'] > start_date) & (new_york_data['Date'] <= end_date)
new_york_new = new_york_data.loc[new2]
new_york_new
```

```python
chicago_data.head()
```

```python
start_date='1/1/2006'
end_date='12/31/2017'
new = (chicago_data['Date'] > start_date) & (chicago_data['Date'] <= end_date)
chicago_new = chicago_data.loc[new]
chicago_new
```

```python
chicago_new['Date'] = pd.to_datetime(chicago_new['Date']).dt.strftime('%m/%d/%Y')
chicago_new.head()
```

```python
# chicago_new['Date']=chicago_new['Date'].dt.strftime('%m/%d/%Y')
# chicago_new.head()
```

```python
chicago_new.sort_values(by='Date', ascending=True, inplace=True)
chicago_new.head()
```

```python
chicago_new[['Date', 'Primary Type', 'Description']].groupby(["Date","Primary Type"]).count().reset_index()
```

```python
new_york_new[['Date', 'Primary Type', 'Description']].groupby(["Date","Primary Type"]).count().reset_index()
```

```python
crime_data = pd.concat([new_york_new, chicago_new])
crime_data.head(100)
#crime_data = pd.concat(new_york_new, chicago_new, how = "inner", on = "Primary Type")
#crime_data.head()
```

```python
crime_data.to_csv("crime_data.csv", encoding = "utf-8", index = False, header = True)
```
